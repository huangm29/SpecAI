```python
import psi4
import numpy as np
import os

os.environ['KMP_DUPLICATE_LIB_OK']='True' # to fix the omp running issue


acetamide = psi4.geometry("""
O    -0.5447	1.3323   -0.0013 
N     0.6985   -0.6359    0.0058 
C    -0.4701	0.1063    0.0070 
C     2.0143   -0.0386   -0.0057 
C    -1.6980   -0.7641   -0.0059 
H     0.6423   -1.6500   -0.0141 
H     2.6279   -0.5471    0.7418 
H     1.9657	1.0301    0.2163 
H     2.4489   -0.1908   -0.9966 
H    -2.5800   -0.1596    0.2234 
H    -1.8162   -1.2127   -0.9959 
H    -1.6122   -1.5549    0.7449 
""")

psi4.optimize('b3lyp/6-31g*', molecule=acetamide)

scf_e, scf_wfn = psi4.frequency('b3lyp/6-31g*', molecule=acetamide, return_wfn=True)
```

```python
freq = scf_wfn.frequency_analysis['omega'].data # frequencies of normal modes
mode = scf_wfn.frequency_analysis['x'].data # normal modes, normalized un-mass-weighted
IR_int = scf_wfn.frequency_analysis['IR_intensity'].data # IR intensity

# print(freq)

vib_fr = [f.real for f in freq[6:]] # vibration frequencies start from the 6th
print(vib_fr)

mode = np.transpose(mode)[6:] # now the row vectors represent the x, y, z of atoms in order
print(mode)

IR_int = IR_int[6:]
print(IR_int)
```

```python
# write the data file

datafile = "simu_data.dat"

with open(datafile,"w") as DATA:
    for i in range(len(vib_fr)):
        DATA.write('%-20f %-20f\n' %(vib_fr[i],IR_int[i]))
```
