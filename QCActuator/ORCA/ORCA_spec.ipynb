# ORCA Spectroscopy Tools - Example

This notebook demonstrates how to run various spectroscopy calculations on a single molecule using the enhanced ORCA tools.

```python
# Import our quantum chemistry framework
from orca_tools import OrcaCalculator, OrcaConfig
```

## 1. Setup ORCA Path

First, define the path to your ORCA installation:

```python
# Update this path to match your ORCA installation
ORCA_PATH = "/home/huangm/source/orca_6_0_1_linux_x86-64_shared_openmpi416/"

# Create a config with verbose output enabled
config = OrcaConfig(verbose=False)

# Create a new calculator
calc = OrcaCalculator(orca_path=ORCA_PATH, config=config)
```

## 2. Define Molecular Geometry

Let's define the molecules for our examples:

```python
# Formaldehyde
formaldehyde= """
  C          -0.00000007544128      0.00000007722154      0.02576945733573
  O           0.00000019009665     -0.00000002553588      1.23240550391654
  H           0.93816203542447     -0.00000002584283     -0.56908732489374
  H          -0.93816215007984     -0.00000002584283     -0.56908763635853
"""
```

```python
# Cyclopropanol
Cyclopropanol= """
  C      1.231429   -0.226472   -0.084960
  C     -0.061893    0.507641    0.134338
  C     -1.358912   -0.147897    0.084831
  O     -0.902881    0.641038   -0.969176
  H      1.070541   -1.118875   -0.689778
  H      1.672013   -0.522768    0.869009
  H      1.946503    0.413187   -0.605194
  H      0.017832    1.411161    0.734623
  H     -1.417896   -1.212878   -0.118068
  H     -2.196737    0.255864    0.644375
"""
```

## 3. Visualize Cyclopropanol

Let's first visualize the Cyclopropanol molecule using py3Dmol:

```python
# Visualize Cyclopropanol
calc.visualize_molecule(geometry=Cyclopropanol)
```

## 4. Run UV-Vis Spectroscopy Calculation

Let's run a UV-Vis absorption spectroscopy calculation on formaldehyde:

```python
# Run UV-Vis calculation for formaldehyde
print("Running UV-Vis calculation...")
uv_spectrum = calc.run_spectroscopy(
    spec_type="UV", 
    geometry=formaldehyde
)
```

## 5. Run ECD Spectroscopy Calculation

Let's run an ECD calculation on formaldehyde:

```python
# Run ECD calculation for Cyclopropanol
print("Running ECD calculation...")
uv_spectrum = calc.run_spectroscopy(
    spec_type="ECD", 
    geometry=Cyclopropanol,
    basis="def2-SVP",0
)
```

## 6. Run O K-edge X-ray Absorption Spectroscopy Calculation

Let's run an XAS calculation on formaldehyde:

```python
# Run O K-edge XAS calculation for formaldehyde
print("Running O K-edge XAS calculation...")
xas_spectrum = calc.run_spectroscopy(
    spec_type="XAS", 
    geometry=formaldehyde,
    core_orb=0,
    start_point=515.0,
    end_point=530.0
)
```

## 7. Run IR Spectroscopy 

Now let's run an IR calculation on Formaldehyde (this will include OPT and FREQ):

```python
# Run IR calculation for acetone (this will automatically add OPT and FREQ keywords)
print("Running IR calculation (with optimization)...")
ir_spectrum = calc.run_spectroscopy(
    spec_type="IR", 
    geometry=formaldehyde
    )
```

## 8. Run Raman Spectroscopy 

Now let's run a Rmaan calculation on Formaldehyde (this will include OPT and NUMFREQ): 

```python
# Run Raman calculation for acetone (this will automatically add OPT and FREQ keywords)
print("Running Raman calculation (with optimization)...")
raman_spectrum = calc.run_spectroscopy(
    spec_type="Raman", 
    geometry=formaldehyde
)
```

## 9. Run VCD Spectroscopy 

Now let's run a VCD calculation on Cyclopropanol (this will include OPT and FREQ):

```python
# Run VCD calculation for acetone (this will automatically add OPT and FREQ keywords)
print("Running VCD calculation (with optimization)...")
ir_spectrum = calc.run_spectroscopy(
    spec_type="VCD", 
    geometry=Cyclopropanol,
    basis="def2-SVP"
    )  
```

## 10. Run NMR Spectroscopy

Let's run an NMR calculation on formaldehyde:

```python
# Run NMR calculation for acetone
print("Running NMR calculation...")
nmr_spectrum = calc.run_spectroscopy(
    spec_type="NMR", 
    geometry=formaldehyde
)
```

## 11. Run EFEI Calculation

Let's run an EFEI calculation on formaldehyde:

```python
data = calc.run_efei_calculation(
    geometry=formaldehyde,
    efei_atoms=(2, 3),  # Field direction from atom 2 to atom 3
    efei_strength=6.0  # Field strength in atomic units
)
```
